# Reading from Shapefile

To successfully read a shapefile using GeoPandas, you typically need the following associated files:

- `.shp`: This is the main file that contains the geometric data (the actual shapes like points, lines, or polygons).

- `.shx`: The shape index file that allows for quick access to the geometric data in the .shp file.

- `.dbf`: The attribute file that contains the attribute data for the geometries, stored in a table format.

- `.prj` (optional but recommended): This file contains the projection information, which defines the coordinate system used by the shapefile.

```python
import geopandas as gpd

# Read the shapefile
gdf_shapefile = gpd.read_file('MajorEarthRivers/MajorRivers.shp')
gdf_shapefile
```

```python
gdf_shapefile['SYSTEM'].unique()
```

Some rivers have a None value in the SYSTEM column, which could mean they are not part of a larger recognized system or the data lacks that information.

```python
gdf_shapefile.plot()
```

# Reading from CSV

**7 continents**:  Asia, Europe, Africa, North America, South America, Australia, Antarctica.

```python
import pandas as pd
import geopandas as gpd

# Load a CSV file with latitude and longitude columns
df = pd.read_csv('airports.csv', na_filter=False)
df.head()
```

```python
# Convert latitude and longitude into Points
geometry = gpd.points_from_xy(df.longitude, df.latitude)

# Create a GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry=geometry)
gdf.head()
```

```python
gdf.shape
```

```python
gdf['type'].value_counts()
```

```python
gdf['continent'].value_counts()
```

**7 continents**:  Asia, Europe, Africa, North America, South America, Australia, Antarctica.

```python
gdf['iso_country'].unique()
```

```python
gdf['iso_country'].nunique()
```

```python
gdf[gdf['iso_country'] == 'US'].iso_country.value_counts()
```

```python
gdf.plot(markersize=0.05)
```
