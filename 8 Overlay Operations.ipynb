# Overlay Operations

```python
import geopandas as gpd

chicago_comm = gpd.read_file('https://geodacenter.github.io/data-and-lab/data/Chi-CCA.zip')
chicago_comm.head()
```

```python
chicago_comm.shape
```

```python
chicago_comm.plot()
```

```python
chicago_comm.crs
```

```python
chicago_sdoh = gpd.read_file('https://geodacenter.github.io/data-and-lab/data/Chi-SDOH.zip')
chicago_sdoh.head()
```

```python
chicago_sdoh.shape
```

```python
chicago_sdoh.plot()
```

```python
chicago_sdoh.crs
```

```python
# Ensure both the datsets are in the same CRS
chicago_sdoh = chicago_sdoh.to_crs(chicago_comm.crs)
```

```python
chicago_sdoh.crs
```

## Intersection

```python
intersection = gpd.overlay(chicago_comm, chicago_sdoh, how='intersection')
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,10))
chicago_comm.plot(ax=ax, color='lightgrey', edgecolor='black', alpha=0.7)
intersection.plot(ax=ax, color='red', edgecolor='black', alpha=0.5)
plt.title('Intersection of Chicago Community Areas and SDOH Tracts')
plt.show()
```

## Union

```python
union = gpd.overlay(chicago_comm, chicago_sdoh, how='union', keep_geom_type=False)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
union.plot(ax=ax, color='forestgreen', edgecolor='indigo', alpha=0.5)
plt.title('Union of Chicago Community Areas and SDOH Tracts')
plt.show()
```

## Difference

```python
difference = gpd.overlay(chicago_comm, chicago_sdoh, how='difference', keep_geom_type=False)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
chicago_comm.plot(ax=ax, color='lightgrey', edgecolor='black', alpha=0.5)
difference.plot(ax=ax, color='orange', edgecolor='black', alpha=0.7)
plt.title('Difference: Chicago Community Areas not in SDOH Tracts')
plt.show()
```

```python
difference2 = gpd.overlay(chicago_sdoh, chicago_comm, how='difference', keep_geom_type=False)

fig, ax = plt.subplots(figsize=(10,10))
chicago_comm.plot(ax=ax, color='lightgrey', edgecolor='black', alpha=0.5)
difference2.plot(ax=ax, color='orange', edgecolor='black', alpha=0.7)
plt.title('Difference: SDOH Tracts not in Chicago Community Areas')
plt.show()
```

## Symmetric Difference

```python
symmetric_difference = gpd.overlay(chicago_comm, chicago_sdoh, how='symmetric_difference', keep_geom_type=False)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
chicago_comm.plot(ax=ax, color='lightgrey', edgecolor='black', alpha=0.5)
symmetric_difference.plot(ax=ax, color='green', edgecolor='black', alpha=0.7)
plt.title('Symmetric Difference: Areas Unique to Each Dataset')
plt.show()
```

```python
chicago_airbnb = gpd.read_file('https://geodacenter.github.io/data-and-lab/data/airbnb.zip')
chicago_airbnb = chicago_airbnb.to_crs(chicago_comm.crs)
chicago_airbnb.plot()
```

```python
symmetric_difference2 = gpd.overlay(chicago_sdoh, chicago_airbnb, how='symmetric_difference', keep_geom_type=False)

fig, ax = plt.subplots(figsize=(10,10))
chicago_comm.plot(ax=ax, color='lightgrey', edgecolor='black', alpha=0.5)
symmetric_difference2.plot(ax=ax, color='green', edgecolor='black', alpha=0.7)
plt.title('Symmetric Difference: Areas Unique to Each Dataset')
plt.show()
```
