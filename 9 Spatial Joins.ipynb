```python
import geopandas as gpd

chicago_comm = gpd.read_file('https://geodacenter.github.io/data-and-lab/data/Chi-CCA.zip')
chicago_comm.head()
```

```python
chicago_supermarket = gpd.read_file('/content/Supermarket/grocery/chicago_sup.shp')
chicago_supermarket.head()
```

```python
chicago_supermarket.plot()
```

```python
chicago_supermarket.crs
```

```python
chicago_comm.crs
```

```python
chicago_supermarket = chicago_supermarket.to_crs(chicago_comm.crs)
```

```python
# Spatial Join with 'within'
within_join = gpd.sjoin(chicago_supermarket, chicago_comm, how='inner', predicate='within')
within_join.head()
```

```python
import matplotlib.pyplot as plt

# Plot Within Spatial Join
fig, ax = plt.subplots(figsize=(10,10))
chicago_comm.plot(ax=ax, color='lightgrey', edgecolor='black', alpha=0.5)
within_join.plot(ax=ax, color='brown', markersize=10, label='Supermarkets (Within)')
plt.title('Spatial Join (Within): Supermarkets within the Community Areas')
plt.legend()
plt.show()
```

```python
# Spatial Join with 'Contains'
contains_join = gpd.sjoin(chicago_comm, chicago_supermarket, how='inner', predicate='contains')
contains_join.head()
```

```python
# Plot Contails Spatial Join
fig, ax = plt.subplots(figsize=(10,10))
chicago_comm.plot(ax=ax, color='lightgrey', edgecolor='black', alpha=0.5)
contains_join.plot(ax=ax, color='green', label='Supermarkets', alpha=0.7)
plt.title('Spatial Join (Contains): Supermarkets in Community Areas')
#plt.legend()
plt.show()
```

```python
# Spatial join with 'Intersects'
intersects_join = gpd.sjoin(chicago_supermarket, chicago_comm, how='inner', predicate='intersects')
intersects_join.head()
```

```python
# Plot Intersects Spatial Join
fig, ax = plt.subplots(figsize=(10,10))
chicago_comm.plot(ax=ax, color='lightgrey', edgecolor='black', alpha=0.5)
intersects_join.plot(ax=ax, color='red', markersize=10, label='Supermarkets (Intersects)', alpha=0.8)
plt.title('Spatial Join (Intersects): Supermarkets intersecting the Community Areas')
plt.legend()
plt.show()
```

```python
right_join = gpd.sjoin(chicago_supermarket, chicago_comm, how='right', predicate='intersects')

# Plot Right Spatial Join
fig, ax = plt.subplots(figsize=(10,10))
#chicago_comm.plot(ax=ax, color='lightgrey', edgecolor='black', alpha=0.5)
right_join.plot(ax=ax, color='purple', edgecolor='black', label='Supermarkets (Right Join)', alpha=0.7)
plt.title('Right Spatial Join: Supermarkets with Community Areas')
#plt.legend()
plt.show()
```

```python
left_join = gpd.sjoin(chicago_supermarket, chicago_comm, how='left', predicate='intersects')

# Plot Left Spatial Join
fig, ax = plt.subplots(figsize=(10,10))
chicago_comm.plot(ax=ax, color='lightgrey', edgecolor='black', alpha=0.5)
left_join.plot(ax=ax, color='blue', markersize=10, label='Supermarkets (Left Join)', alpha=0.7)
plt.title('Left Spatial Join: Supermarkets with Community Areas')
plt.legend()
plt.show()
```
